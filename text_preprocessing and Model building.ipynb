# Here we clearly observe that data have some non-alphabets,special symbols,numbers etc

# In text preprocessing we do removing stopwords,conevrting data into lowercase,removing whitespaces,replacing non-alphabets,stemming the word...

# we use natural language tool kit for doing preprocessing of text data

# Stemming is the process of reducing a word to its word stem that affixes to suffixes and prefixes or to the roots of words known as a lemma. Stemming is important in natural language understanding (NLU) and natural language processing (NLP).

# Lemmatization usually refers to doing things properly with the use of a vocabulary and morphological analysis of words, normally aiming to remove inflectional endings only and to return the base or dictionary form of a word, which is known as the lemma .

```python
import re
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from string import punctuation
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
lemmatizer = WordNetLemmatizer()
from textblob import TextBlob
X=[]
for i in range(0,30000):
       comment=toxic["comment_text"][i]
       comment = re.sub('[^a-zA-Z]', ' ', comment)
       comment = comment.lower()
       comment= comment.split()
       comment = [ps.stem(word) for word in comment if not word in set(stopwords.words('english'))]
       comment = ' '.join(comment)
       X.append(comment)
```

```python
X[:100]
```

# Tokenization

# Tokenization is essentially splitting a phrase, sentence, paragraph, or an entire text document into smaller units, such as individual words or terms.

```python
from nltk.tokenize import sent_tokenize, word_tokenize 
token=[]
for i in X:
    token.append(word_tokenize(i))
    
```

# Bag of words

```python
        
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=1500,stop_words='english')
X1 = cv.fit_transform(X).toarray()
```

```python
ip_dim = X1.shape[1]
```

```python
y = toxic_labels.values
```

# Train and test split the data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X1, y, test_size=0.20, random_state=42)
```

# ANN Model

# Importing Model building libraries

```python
from keras.models import Sequential
from keras.layers import Dense
```

# Adding input layer,hidden layer and output layer

```python
model = Sequential()

model.add(Dense(output_dim = 64, init = 'uniform', activation = 'relu', input_dim = ip_dim))#input layer

model.add(Dense(output_dim = 32, init = 'uniform', activation = 'relu'))#hidden layer

model.add(Dense(output_dim = 6, init = 'uniform', activation = 'sigmoid'))#output layer
```

# Compliling the model

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

# Train and Test the model 


```python
model.fit(X_train,y_train,batch_size=32,epochs=50)
```

# Here we use optimizer as "adam",loss function as binary_crossentropy

```python
y_pred=model.predict(X_test)

y_pred=y_pred>0.5
```

```python
y_pred
```

```python
y_test
```

```python
score = model.evaluate(X_test, y_test)

print("Test Score:", score[0])
print("Test Accuracy:", score[1])
```
